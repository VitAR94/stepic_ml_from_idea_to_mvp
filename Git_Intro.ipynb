## Системы контроля версий

**Системы контроля версий**, СКВ (**Version Control System**, VCS) — программное обеспечение для облегчения работы с изменяющейся информацией.

**Какую проблему решают?**  

*   Контроль кода: мгновенный откат и отмена изменений при необходимости
*   Контроль действий: все "на виду"
*   Резервное копирование исходного кода для удаленного хранения
*   Простое добавление новых "фич" в код проекта
*   Разрешение конфликтов

СКВ дисциплинирует программиста и придаёт устойчивость проекту.  
Так получается, что даже если вы ведет разработку проекта в одиночку, без нее не обойтись!

## Git

### Основные принципы



1.   Полная копия репозитория хранится локально  
     + все работает очень быстро
     + полный контроль над репозиторием
     + все внутренние файлы Git хранит только в одной папке `.git`
     + можно работать без подключения к интернету

2.   Мыслим коммитами, а не файлами!
     + Каждый раз когда делаем коммит (фиксируем изменения), то указываем какие именно *изменения* хотим в него добавить
     + можно добавить в коммит только часть внесенных изменений в файле, а другую часть сделать отдельным коммитом

3.   Контроль над коммитами
     + удаление
     + редактирование
     + объединениие
     + разбиение одного коммита на несколько других

3.   Ветки
     + создание
     + переключение между ветками
     + слияние
     + удаление
